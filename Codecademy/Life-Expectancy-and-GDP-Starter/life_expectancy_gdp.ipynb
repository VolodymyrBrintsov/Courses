```python
sns.set_style('white')
sns.set_palette('Set3')
plt.figure(figsize=(15, 8))
sns.barplot(data=df, x='Country', y='GDP')
plt.title('GDP by Country')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=df, x='Country', y='Life expectancy at birth (years)')
plt.title('Life expectancy at birth (years) by Country')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15, 8))
countries = df['Country'].unique()
for i, country in enumerate(countries):
    ax = plt.subplot(2, 3, i+1)
    data = df[df['Country'] == country]
    plt.plot(data['Year'], data['GDP'], label=country, marker='o', color='red')
    plt.title(f'GDP changing by {country}')
    plt.xlabel('Year')
    plt.ylabel('GDP')
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.show()
```
